```python
import numpy as np
from scipy.stats import norm

def black_scholes(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + (sigma**2)/2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    # Calcul du prix du Call européen
    Call = S * norm.cdf(d1) - K * np.exp(- r * T) * norm.cdf(d2)

    # Calcul du prix du Put européen 
    Put = K * np.exp(- r * T ) * norm.cdf(-d2) - S * norm.cdf(-d1)

    
    # Calcul des Greeks : 

    # Calcul du Delta 
    delta_Call = norm.cdf(d1)
    delta_Put = norm.cdf(d1) - 1 

    # Calcul du Gamma
    Gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))

     # Calcul du Vega
    Vega = S * norm.pdf(d1) * np.sqrt(T)

    # Calcul du Theta
    Theta_Call = - (S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))) - r * K * np.exp(- r * T ) * norm.cdf(d2)
    Theta_Put = - (S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))) + r * K * np.exp(- r * T) * norm.cdf(-d2)
    
    # Calcul du Rho 
    Rho_Call = K * T * np.exp(-r * T) * norm.cdf(d2)
    Rho_Put = - K * T * np.exp(-r * T) * norm.cdf(-d2)

    return Call, Put, delta_Call, delta_Put, Gamma, Vega, Theta_Call, Theta_Put, Rho_Call, Rho_Put
    
    

Call, Put, delta_Call, delta_Put, Gamma, Vega, Theta_Call, Theta_Put, Rho_Call, Rho_Put = black_scholes(100, 100, 1, 0.05, 0.20)
print(f"Call : {Call:.4f}")
print(f"Put : {Put:.4f}")
print(f"Delta du Call : {delta_Call:.4f}")
print(f"Delta du Put : {delta_Put:.4f}")
print(f"Gamma : {Gamma:.4f}")
print(f"Vega : {Vega:.4f}")
print(f"Theta du Call : {Theta_Call:.4f}")
print(f"Theta du Put : {Theta_Put:.4f}")
print(f"Rho du Call : {Rho_Call:.4f}")
print(f"Rho du Put : {Rho_Put:.4f}")
```
