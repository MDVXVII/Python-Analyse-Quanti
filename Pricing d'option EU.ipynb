```python
import numpy as np
from scipy.stats import norm

def black_scholes(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + (sigma**2)/2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    # Calcul du prix du Call européen
    Call = S * norm.cdf(d1) - K * np.exp(- r * T) * norm.cdf(d2)

    # Calcul du prix du Put européen 
    Put = K * np.exp(- r * T ) * norm.cdf(-d2) - S * norm.cdf(-d1)

    
    # Calcul des Greeks : 

    # Calcul du Delta 
    delta_Call = norm.cdf(d1)
    delta_Put = norm.cdf(d1) - 1 

    # Calcul du Gamma
    Gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))

    # Calcul du Vega
    Vega = S * norm.pdf(d1) * np.sqrt(T)

    # Calcul du Theta
    Theta_Call = - (S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))) - r * K * np.exp(- r * T ) * norm.cdf(d2)
    Theta_Put = - (S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))) + r * K * np.exp(- r * T) * norm.cdf(-d2)

    # Calcul du Rho 
    Rho_Call = K * T * np.exp(-r * T) * norm.cdf(d2)
    Rho_Put = - K * T * np.exp(-r * T) * norm.cdf(-d2)

    return {"call" : Call, 
            "put" : Put, 
            "delta_call" : delta_Call, 
            "delta_put" : delta_Put, 
            "gamma" : Gamma, 
            "vega" : Vega, 
            "theta_call" : Theta_Call, 
            "theta_put" : Theta_Put, 
            "rho_call" : Rho_Call, 
            "rho_put" : Rho_Put 
        }
    
# Afficher les resultats 
resultats = black_scholes(100, 100, 1, 0.05, 0.20)
for nom, valeur in resultats.items():
    print(f"{nom} : {valeur:.4f}")
```
